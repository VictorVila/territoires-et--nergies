```python
import pandas as pd
pd.set_option('display.max_columns', 1000)
```

```python
# assesment = ass 
ass = pd.read_csv('data/ADEME/beges/assessments.csv', dtype={'id': str})
ass.head(3)
```

```python
ass.collectivity_type.unique()
```

```python
ass=ass.loc[ass['collectivity_type'].notnull()]
```

```python
legal=pd.read_csv('data/ADEME/beges/legal_units.csv', dtype=str)
legal.head(3)
```

```python
df1=pd.merge(legal, ass, how='inner', left_on='assessment_id', right_on='id')
df1.head(3)
```

```python
df1.reporting_year.unique()
```

```python
df1.shape
```

```python
emm=pd.read_csv('data/ADEME/beges/emissions.csv', dtype={'assessment_id': str})
emm.head(3)
```

```python
emm.shape
```

```python
df2=pd.merge(df1, emm, how='inner', left_on='assessment_id', right_on='assessment_id')
df2.head(3)
```

```python
df2.shape
```

```python
scope = pd.read_csv('data/ADEME/beges/scope_items.csv')
scope.head()
```

```python
df3=pd.merge(df2, scope, how='left', left_on='scope_item_id', right_on='scope_id' )
df3.head()
```
