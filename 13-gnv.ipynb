### EPCI

```python
import pymysql.cursors

connection = pymysql.connect( host='localhost',
                              user='root',
                              passwd='123',
                              db='energie',
                              charset='utf8mb4',
                              cursorclass=pymysql.cursors.DictCursor)

cursor = connection.cursor()
```

```python
epci_data = pd.read_csv('data/insee/communes_cp_insee_epci_geo.csv', dtype=str)
epci_data = epci_data.drop_duplicates()
epci_data.head(3)
```

```python
epci_data = epci_data.drop([ 'dep', 'reg'], axis=1)
epci_data.head(3)
```

```python
#epci 

epci = pd.merge(df, epci_data, how='left', left_on="Code INSEE commune", right_on='codgeo')
```

```python
epci=epci.drop_duplicates()
epci.shape
```

```python
epci=epci.dropna(how='any')
epci.shape
```

```python
# Insertion BDD 

with connection.cursor() as cursor:
    
    sql = ("INSERT INTO `gnv` "
           "(code_insee, epci, exploitant, ad_station, Xlongitude, Ylatitude, "
           " carburant, bio_gnc, statut, acces_pl)"
           " VALUES (%s, %s, %s, %s, %s, %s, "
           " %s, %s, %s, %s)"
          )
    #print(sql)
    
    for i, r in epci.iterrows():
        ad_station = r['Rue'] + ' - ' + r['codepostal'] + ' - ' + r['libcommune']
        cursor.execute(sql, 
            ( 
            r['Code INSEE commune'],
            r['epci'],
            r['Exploitant'],
            ad_station,
            r['Longitude station (DD)'],
            r['Latitude station (DD)'],
            r['Carburant'],
            r['Bio-GNC'],
            r['Statut'],
            r['Accès PL'] 
            )
        )

connection.commit() 
```
